# hello_node — green run (notebook)

A dependency-free NeuroWorkflow example. **Run all cells**, then compare the
printed numbers with `EXPECTED_OUTPUT.md`. Run this notebook from the
`hello_node/` folder so the node files are importable.

```python
import os, sys

_HERE = os.getcwd()  # run this notebook from the hello_node/ folder
if _HERE not in sys.path:
    sys.path.insert(0, _HERE)

from neuroworkflow import WorkflowBuilder
from signal_generator import SignalGeneratorNode
from signal_statistics import SignalStatisticsNode

generator = SignalGeneratorNode("Generator")
statistics = SignalStatisticsNode("Stats")
generator.configure(length=100, amplitude=1.0, cycles=2.0)

wf = WorkflowBuilder("HelloNodeWorkflow")
wf.add_node(generator)
wf.add_node(statistics)
wf.connect("Generator", "signal", "Stats", "signal")
workflow = wf.build()

success = workflow.execute()
print("workflow success:", success)
```

```python
mean = statistics.get_output("mean")
std = statistics.get_output("std")
n = statistics.get_output("n_samples")
print("mean:", mean)
print("std:", std)
print("n_samples:", n)

# Golden checks — these fail loudly if a port silently came back None / wrong.
assert success is True
assert n == 100
assert abs(mean) < 1e-9          # full sine period averages to ~0
assert abs(std - 0.7071067811865476) < 1e-9
print("\nOK — matches EXPECTED_OUTPUT.md")
```

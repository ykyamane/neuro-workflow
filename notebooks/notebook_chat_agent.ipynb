# NeuroWorkflow Notebook Chat Agent

An AI chat agent that runs inside this Jupyter kernel (Issue #52).

- The agent loop runs in the kernel, so it can run code and edit files directly in your workspace.
- The OpenAI key and the workflow (MCP) tools stay on the backend; the kernel talks to it over HTTP.
- It reads the skills tracked in the repo's `.claude/skills/` and uses them as guidance.

Requires the JupyterHub spawner wiring (backend URL + service token), which is set up automatically in the Docker stack.

## Option A — inline magic

```python
%load_ext neuroworkflow.agent
```
